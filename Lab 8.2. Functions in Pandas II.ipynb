## 1. Let's load some textual data

```python
url = 'https://raw.githubusercontent.com/thousandoaks/BEMM458/master/data/realdonaldtrump.csv'
TrumpTweets = pd.read_csv(url, index_col=0)
```

## 2. Extract URLs from the text

```python
pd.options.display.max_colwidth = 200
TrumpTweets['content'].sample(3)
```

```python
TrumpTweets['url']=TrumpTweets['content'].apply(lambda x:re.findall(r'(https?://\S+)', x))
```

```python
TrumpTweets[['content','url']].sample(10)
```

## 3. Extract References from the text

```python
TrumpTweets['hashtags_extracted']=TrumpTweets['content'].apply(lambda x:re.findall(r'\B(\#[a-zA-Z]+\b)(?!;)'    , x))
```

```python
TrumpTweets[['content','hashtags','hashtags_extracted']].head(30)
```
